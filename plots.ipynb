```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#    
data = pd.read_csv("5restarts1%.csv", sep=';', header=0)

#    
methods = data['type'].unique()

#     (train)   (test)
fig_train, axs_train = plt.subplots(1, 2, figsize=(20, 10))
fig_test, axs_test = plt.subplots(1, 2, figsize=(20, 10))

for method in methods:
    #         
    method_data = data[data['type'] == method].sort_values(by='epoch')
    
    #   ,     
    grouped = method_data.groupby('epoch')
    epochs = sorted(grouped.groups.keys())
    
    #          
    train_log_mean = grouped['train_log'].mean().values
    train_log_std = grouped['train_log'].std().values
    train_acc_mean = grouped['train_acc'].mean().values
    train_acc_std = grouped['train_acc'].std().values

    test_log_mean = grouped['test_log'].mean().values
    test_log_std = grouped['test_log'].std().values
    test_acc_mean = grouped['test_acc'].mean().values
    test_acc_std = grouped['test_acc'].std().values

    #    loss  accuracy  
    axs_train[0].plot(epochs, train_log_mean, label=method)
    axs_train[0].fill_between(epochs, train_log_mean - train_log_std, train_log_mean + train_log_std, alpha=0.3)
    
    axs_train[1].plot(epochs, train_acc_mean, label=method)
    axs_train[1].fill_between(epochs, train_acc_mean - train_acc_std, train_acc_mean + train_acc_std, alpha=0.3)
    
    #    loss  accuracy  
    axs_test[0].plot(epochs, test_log_mean, label=method)
    axs_test[0].fill_between(epochs, test_log_mean - test_log_std, test_log_mean + test_log_std, alpha=0.3)
    
    axs_test[1].plot(epochs, test_acc_mean, label=method)
    axs_test[1].fill_between(epochs, test_acc_mean - test_acc_std, test_acc_mean + test_acc_std, alpha=0.3)

#        
axs_train[0].set_title("Train Loss")
axs_train[0].set_xlabel("Epoch")
axs_train[0].set_ylabel("Loss")
axs_train[0].legend()
axs_train[0].grid()

axs_train[1].set_title("Train Accuracy")
axs_train[1].set_xlabel("Epoch")
axs_train[1].set_ylabel("Accuracy")
axs_train[1].legend()
axs_train[1].grid()

#        
axs_test[0].set_title("Test Loss")
axs_test[0].set_xlabel("Epoch")
axs_test[0].set_ylabel("Loss")
axs_test[0].legend()
axs_test[0].grid()

axs_test[1].set_title("Test Accuracy")
axs_test[1].set_xlabel("Epoch")
axs_test[1].set_ylabel("Accuracy")
axs_test[1].legend()
axs_test[1].grid()

plt.show()
```
